```python
import numpy as np
import pandas as pd
```

# Table Creation

```python
data = pd.read_csv("../data/CleanedData.csv",index_col='Unnamed: 0')
data.head()
```

```python
pivot_table_year = pd.pivot_table(
    data,
    index="Year",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
)
pivot_table_year
```

```python
pivot_table_month = pd.pivot_table(
    data,
    index="Month",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
).reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
pivot_table_month
```

```python
pivot_table_coo = pd.pivot_table(
    data,
    index="Country_of_origin",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
)
pivot_table_coo
```

```python
data04_08 = data[data.Year.between(2004,2008)]
data09_13 = data[data.Year.between(2009,2013)]
data14_18 = data[data.Year.between(2014,2018)]
data19_22 = data[data.Year.between(2019,2022)]
```

# Visualisations

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig,axes = plt.subplots(1,2,figsize=(10,5),sharey=True)

sns.barplot(
    ax = axes[0],
    x=pivot_table_coo["Imports_hardcoal_tonnes"],
    y=pivot_table_coo.index,
    linewidth=1,
    edgecolor="k"
    ).set(xlabel="Tonnes",ylabel="Country of Origin")

sns.barplot(
    ax = axes[1],
    x=pivot_table_coo["Imports_hardcoal_eurospert"],
    y=pivot_table_coo.index,
    linewidth=1,
    edgecolor="k"
).set(xlabel="Euros Per Tonne",ylabel=None)

plt.tight_layout()
```
