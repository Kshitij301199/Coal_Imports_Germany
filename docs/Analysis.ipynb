```python
import numpy as np
import pandas as pd
```

# Table Creation

```python
writer = pd.ExcelWriter("../data/Pivot_Tables.xlsx",engine='xlsxwriter')
```

```python
data = pd.read_csv("../data/CleanedData.csv",index_col='Unnamed: 0')
data.head()
```

```python
pivot_table_year = pd.pivot_table(
    data,
    index="Year",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
)
pivot_table_year.to_excel(writer,sheet_name="Year")
pivot_table_year
```

```python
pivot_table_month = pd.pivot_table(
    data,
    index="Month",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
).reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
pivot_table_month.to_excel(writer,sheet_name="Month")
pivot_table_month
```

```python
pivot_table_coo2 = pd.pivot_table(
    data,
    index="Country_of_origin",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.sum,
        "Imports_hardcoal_euros":np.sum,
        "Imports_hardcoal_eurospert":np.sum
    }
)
```

```python
pivot_table_coo = pd.pivot_table(
    data,
    index="Country_of_origin",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
)
pivot_table_coo.to_excel(writer,sheet_name="Country of Origin")
pivot_table_coo
```

```python
writer.close()
```

```python
data04_08 = data[data.Year.between(2004,2008)]
data09_13 = data[data.Year.between(2009,2013)]
data14_18 = data[data.Year.between(2014,2018)]
data19_22 = data[data.Year.between(2019,2022)]
```

```python
pivot_table_month0408 = pd.pivot_table(
    data04_08,
    index="Month",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
).reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
# pivot_table_month
pivot_table_month0913 = pd.pivot_table(
    data09_13,
    index="Month",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
).reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
# pivot_table_month
pivot_table_month1418 = pd.pivot_table(
    data14_18,
    index="Month",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
).reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
# pivot_table_month
pivot_table_month1922 = pd.pivot_table(
    data19_22,
    index="Month",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
).reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
# pivot_table_month
```

```python
mean_imports_tonnes_pm = []
for i,j,k,l in zip(pivot_table_month0408['Imports_hardcoal_tonnes'],pivot_table_month0913['Imports_hardcoal_tonnes'],pivot_table_month1418['Imports_hardcoal_tonnes'],pivot_table_month1922['Imports_hardcoal_tonnes']):
    mean_imports_tonnes_pm.append(np.mean([i,j,k,l]))
```
