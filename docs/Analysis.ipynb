```python
import numpy as np
import pandas as pd
```

# Table Creation

```python
data = pd.read_csv("../data/CleanedData.csv",index_col='Unnamed: 0')
data.head()
```

```python
pivot_table_year = pd.pivot_table(
    data,
    index="Year",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
)
pivot_table_year
```

```python
pivot_table_month = pd.pivot_table(
    data,
    index="Month",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
).reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
pivot_table_month
```

```python
pivot_table_coo = pd.pivot_table(
    data,
    index="Country_of_origin",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
)
pivot_table_coo
```

```python
data04_08 = data[data.Year.between(2004,2008)]
data09_13 = data[data.Year.between(2009,2013)]
data14_18 = data[data.Year.between(2014,2018)]
data19_22 = data[data.Year.between(2019,2022)]
```

```python
pivot_table_month0408 = pd.pivot_table(
    data04_08,
    index="Month",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
).reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
# pivot_table_month
pivot_table_month0913 = pd.pivot_table(
    data09_13,
    index="Month",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
).reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
# pivot_table_month
pivot_table_month1418 = pd.pivot_table(
    data14_18,
    index="Month",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
).reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
# pivot_table_month
pivot_table_month1922 = pd.pivot_table(
    data19_22,
    index="Month",
    values=["Imports_hardcoal_tonnes","Imports_hardcoal_euros","Imports_hardcoal_eurospert"],
    aggfunc={
        "Imports_hardcoal_tonnes":np.mean,
        "Imports_hardcoal_euros":np.mean,
        "Imports_hardcoal_eurospert":np.mean
    }
).reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
# pivot_table_month
```

```python
mean_imports_tonnes_pm = []
for i,j,k,l in zip(pivot_table_month0408['Imports_hardcoal_tonnes'],pivot_table_month0913['Imports_hardcoal_tonnes'],pivot_table_month1418['Imports_hardcoal_tonnes'],pivot_table_month1922['Imports_hardcoal_tonnes']):
    mean_imports_tonnes_pm.append(np.mean([i,j,k,l]))
```

# Visualisations

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig,axes = plt.subplots(1,2,figsize=(15,7),sharey=True)

sns.barplot(
    ax = axes[0],
    x=pivot_table_coo["Imports_hardcoal_tonnes"],
    y=pivot_table_coo.index,
    linewidth=1,
    edgecolor="k"
    ).set(xlabel="Tonnes",ylabel="Country of Origin")

sns.barplot(
    ax = axes[1],
    x=pivot_table_coo["Imports_hardcoal_eurospert"],
    y=pivot_table_coo.index,
    linewidth=1,
    edgecolor="k"
).set(xlabel="Euros Per Tonne",ylabel=None)

plt.tight_layout()
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(
    data=pivot_table_year,
    x = pivot_table_year.index,
    y = "Imports_hardcoal_tonnes",
)
plt.xticks(pivot_table_year.index[::2]);
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(
    data = pivot_table_month0408,
    x = pivot_table_month0408.index,
    y = "Imports_hardcoal_eurospert",
    color = "r",
    linewidth=2
)
sns.lineplot(
    data = pivot_table_month0913,
    x = pivot_table_month0913.index,
    y = "Imports_hardcoal_eurospert",
    color = "y",
    linewidth=2
)
sns.lineplot(
    data = pivot_table_month1418,
    x = pivot_table_month1418.index,
    y = "Imports_hardcoal_eurospert",
    color = "g",
    linewidth=2
)
sns.lineplot(
    data = pivot_table_month1922,
    x = pivot_table_month1922.index,
    y = "Imports_hardcoal_eurospert",
    color = "b",
    linewidth=2
)
plt.xticks(pivot_table_month0408.index,rotation=90);
plt.legend(["2004-2008","2009-2013","2014-2018","2019-2022"],loc="best");
```
